```python
plt.figure(figsize=(4.5, 3))
groups = df.groupby('Personal Loan')
for name, group in groups:
    plt.plot(group['Income'], group['CCAvg'], label='Accept' if name == 1 else 'Not accept', marker='.', linestyle='')

plt.title('Income and Credit Card Spending by Personal Loan Acceptance')
plt.ylabel('Average Credit Card Spending')
plt.xlabel('Income')
plt.legend()
plt.show()
```

3. A successful financial advisor can be considered one who sells most number of products given the number of clients they consult. Based on this definition, who are your top 10 most successful financial advisors? Follow the steps below:
    - Calculate the number of products sold to each customer. Hint: Convert data type to enable sum of categorical columns.
    - Calculate the number of products sold by each advisor. Hint: Utilize `groupby` on appropriate columns and `sum` to get total per advisor.
    - Calculate the number of customers consulted by each advisor. Hint: Same as last step but use `count` instead of `sum`.
    - Divide number of products sold by number of customer consulted for each advisor.
    - Pick the top 10 advisors based on the conversion rate calculated.

```python
for i in ['Brokerage Account','GIC Account' , 'Online' , 'CreditCard' , 'Personal Loan']:
    df[i] = df[i].astype(int)

df['products sold'] = df[['Brokerage Account','GIC Account' , 'Online' , 'CreditCard' , 'Personal Loan']].sum(axis=1)

products_sold_by_each_advisor = df.groupby('Advisor Name')['products sold'].sum()

n_customers_consulted_by_each_advisor = df.groupby('Advisor Name')['products sold'].count()

top10 = (products_sold_by_each_advisor/n_customers_consulted_by_each_advisor).sort_values(ascending = False).head(10)


print(top10)

conversionR = (products_sold_by_each_advisor/n_customers_consulted_by_each_advisor)
```

4. How many of the top10 advisors have CFP designation? Follow the steps below:

    - Ensure the top 10 advisors from previous questions is saved as DataFrame.
    - Use the Advisor Name values to map with CFP values in the dictionary created above. Hint: Look up `map` method.

```python
# write your answer to #4 here
dicfp = dict(zip(df['Advisor Name'], df['CFP']))
DataFrameTop10 = pd.DataFrame({'Advisor Name': top10.index, 'conversion rate': top10.values})


DataFrameTop10['CFP'] = DataFrameTop10['Advisor Name'].map(dicfp)
n_cfp = DataFrameTop10['CFP'].sum()
print(DataFrameTop10)
print(f'number of the top10 advisors have CFP designation:{n_cfp}.')
```
